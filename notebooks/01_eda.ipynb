# 01 — Exploratory Data Analysis

Explore the raw and processed match data to understand:
- Champion pick frequency distribution
- Win-rate by champion / role
- Draft order patterns
- Synergy and counter relationships


```python
import sys
sys.path.insert(0, '..')  # repo root

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.preprocess import load_processed

sns.set_theme(style='whitegrid')
df = load_processed()
print(df.shape)
df.head()
```

```python
# Champion pick frequency
pick_counts = df['champion_id'].value_counts().head(30)
fig, ax = plt.subplots(figsize=(14, 5))
pick_counts.plot(kind='bar', ax=ax)
ax.set_title('Top 30 Most Picked Champions')
ax.set_xlabel('Champion ID')
ax.set_ylabel('Pick Count')
plt.tight_layout()
plt.show()
```

```python
# Win-rate per champion (require ≥ 50 picks)
win_rate = (
    df.groupby('champion_id')['team_win']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'win_rate', 'count': 'picks'})
    .query('picks >= 50')
    .sort_values('win_rate', ascending=False)
    .head(20)
)
win_rate
```

```python
# Pick order analysis: does pick order correlate with win-rate?
order_wr = df.groupby('pick_order')['team_win'].mean()
order_wr.plot(kind='bar', title='Win Rate by Pick Order (0=first pick)', ylabel='Win Rate')
plt.tight_layout()
plt.show()
```

```python
# Role distribution
df['role'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Role Distribution')
plt.ylabel('')
plt.show()
```
